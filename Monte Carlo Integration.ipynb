# Monte Carlo Integration 

# The following integral is solved using MC Integration technique

```python
#Importing Libraries 
import numpy as np
import statistics as st
from matplotlib import pyplot as plt
```

```python
n = [100, 1000, 10000, 100000] #Number of runs to be used 
H = np.zeros(4) #array that holds the final approximation of all the runs 

# Simulation 
for x in range (4):  
    J= np.linspace (0,1,n[x])
    h=0.00
    for i in J:
        h += (i*np.exp(-(i**2)/2))
    H[x] = (h/n[x])

print(H)
```

```python
#Error Calculation 

Error_arr = np.zeros(4)
exact = 0.39346934
for i in range (4):
    Error_arr[i] = np.abs(H[i]-(exact))
```

```python
#Plotting Error

ax = plt.plot(n, Error_arr)
plt.savefig('Error Medium',dpi=300)
plt.suptitle('Error VS # Runs', fontsize=20)
plt.xlabel('# of Runs', fontsize=18)
plt.ylabel('Error', fontsize=18)
plt.subplots_adjust(top=0.88)
```
